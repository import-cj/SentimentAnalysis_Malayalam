```python
import pandas as pd
from mlmorph import Analyser
from ctypes import *
import re

def mlwordPreprocess(w):
    pat = re.compile(r'[^\u0d00-\u0d7f\s]',re.UNICODE)
    return re.sub(pat, " ", w).strip()

def getstemmer():
 _stemit = WinDLL("libstm.dll").stem_it
 def stemit(s):
  s=mlwordPreprocess(s)
  name = create_string_buffer(s.encode('utf-8'),1024)
  _stemit(name)
  return name.value.decode('utf-8')
 return stemit

def mlmorphStemmer():
    analyser = Analyser()
    st
    def stemmer(s):
        s=mlwordPreprocess(s)
        res = analyser.analyse(s)
        #print(res)
        if not res:
            return s
        res = re.split('<|>', res[0][0])
        i = 0
        rs = ""
        while i < len(res)-1:
            if res[i] == '' or  res[i][0] < u'\u0d00' or res[i][0] > u'\u0d7f':
                i+=2
                continue
            rs += res[i]+"-_-"
            i+=2
        return rs[:-1]
    return stemmer



'''
mlmorphStemmer()("ഇന്ത്യയിൽ")
mlmorphStemmer()("നന്നായിട്ടുണ്ട്")
mlmorphStemmer()("പണത്തിനൊപ്പമുള്ള")
mlmorphStemmer()("(ഇന്ത്യയിൽ)")
'''
stemmer = mlmorphStemmer()
print(stemmer("തോന്നിയിട്ടുണ്ട്"))
print(stemmer("തോന്നിയിട്ടില്ല"))
print(stemmer("ഭൈബ൪എത്രലയ൪ഇടു൦ഏതുന൩൪"))
print(stemmer("ഭആദ്യകാല യഥാർത്ഥ ബ്രാഹ്മണർ ഋഷി മഹർഷി മാരായ ഗണകർ മാത്രമാണ് അവർക്കു ജാതിയും വേണ്ടാ മതവും വേണ്ടാ അവർക്കു ജ്ഞാനവും ലോക ത്തിനു നൽകുക മാത്രമേ വേണ്ടൂ"))
#wdf=pd.read_csv("libvarnam_stem.txt", sep=' ', encoding='utf-8',header=None,names=['t','s'])
#with open("mlmorph_stem.txt", 'w', encoding='utf-8') as fp:
#    for w in wdf['t']:
#        print(w, stemmer(w), file=fp)
```

# Code for Interactive Manual Labelling 

```python
from IPython.display import clear_output
import pandas as pd
import numpy as np
visited = []
startfrom=3
stop=False
def updatefunc(words, s, o, index):
    def labelme(row):
        nonlocal words
        nonlocal s
        nonlocal o
        nonlocal index
        if row.name >= index:
            for word in words:
                if word in row['text']:
                    row['sentiment-label']=s
                    row['offensive-label']=o
                    break
        return row
    return labelme
nm = "sneut.csv"
df = pd.read_csv(nm, sep='\t', encoding='utf-8')
#df["sentiment-label"] = ""
#df["offensive-label"] = ""
for i in range(startfrom, len(df)):
    error=0
    while True:
        try:
            print("-----------------------"+str(i)+"----------------")
            sl = ['sentiment-label', 'offensive-label', "pysent_sent", "pysent_emot", "pysent_hate"]
            s = ''
            for ss in sl:
                s += str(df.iloc[i][ss])+" "
            print(df.iloc[i]["text"])
            print(df.iloc[i]["Entext"])
            a = input("label : "+s)
            a = a.strip()
            if a=='':
                clear_output()
                break
            if a.startswith("x"):
                a = input("word : ")
                a = a.strip()
                v = input("value : ")
                v = v.strip().split()
                df=df.apply(updatefunc(a.split(","), v[0], v[1], i), axis = 1)
                clear_output()
                continue
            if a.startswith("stop"):
                df.to_csv(nm, sep='\t', encoding='utf-8', index=False)
                stop=True
                break
            a=a.split()
            if len(a) == 1:
                a.append("0")
            df.at[i, 'sentiment-label'], df.at[i, 'offensive-label'] = a
            clear_output()
        except:
            df.to_csv(nm, sep='\t', encoding='utf-8', index=False)
            raise
        break
    if i%10==0:
        df.to_csv(nm, sep='\t', encoding='utf-8', index=False)
    if stop:
        break
    visited.append(i)
df.to_csv(nm, sep='\t', encoding='utf-8', index=False)
with open("visited.txt", 'w', encoding='utf-8') as fp:
 print(",".join(map(str,visited)),file=fp)
```
